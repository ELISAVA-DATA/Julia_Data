```python
import pandas as pd
import numpy as np
```

```python
url = "paradas.csv" # fuente: open data barcelona 
    
bbdd = pd.read_csv(url, sep = ',')
```

```python
bbdd.head()
```

```python
bbdd.columns
```

```python
#la columna addresses_neighborhood_name parece contener información sobre los barrios
```

```python
# Agrupar por barrio (addresses_neighborhood_name) y contar las paradas
paradas_por_barrio = bbdd.groupby('addresses_neighborhood_name')['register_id'].count().reset_index(name='cantidad_paradas')

# Ordenar el resultado por cantidad de paradas en orden descendente
paradas_por_barrio = paradas_por_barrio.sort_values(by='cantidad_paradas', ascending=False)

# Mostrar el resultado en la consola
print(paradas_por_barrio)

```
