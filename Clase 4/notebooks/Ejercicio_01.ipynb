```python
import matplotlib.pyplot as plt
```

```python

figura, ax = plt.subplots()

# Definimos las variables
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]
dias = [31,28,31,30,31,30,31,31,30,31,30,31]

# Asigna colores según el número de días
colores = ['blue' if d < 31 else 'orange' for d in dias]

# Ploteamos
ax.bar(meses, dias, color=colores)

# Ajustes eje vertical
ax.set_ylabel('Días',  fontsize=14)

# Ajustes título
ax.set_title('Días de cada mes del año',  fontsize=14)

# Ajustes del eje horizontal
ax.set_xticks(range(len(meses)))  # Asegura que las etiquetas coincidan con las barras
ax.set_xticklabels(meses, rotation=45, fontsize=12)  # Gira las etiquetas y aumenta el tamaño de la fuente

# Ajusta los márgenes para más espacio en el eje X
plt.subplots_adjust(bottom=0.5)  # Incrementa el espacio inferior para las etiquetas

plt.show()
```
